```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams["animation.html"] = "jshtml"
from matplotlib.animation import FuncAnimation
from IPython.display import Video
```

```python
def sine_wave(x, amplitude, frequency, time):
  return amplitude * np.sin(frequency * x + time)
```

```python
fig, ax = plt.subplots()
ax.set_xlim(0, 2 * np.pi)
ax.set_ylim(-1, 1)
ax.set_xlabel('x')
ax.set_ylabel('y')

# Create the line object
line, = ax.plot([], [], lw=2)
```

```python
# Define the update function
def update(frame):
  # Generate the new sine wave data
  x = np.linspace(0, 2 * np.pi, 201)
  y = sine_wave(x, amplitude=x, frequency=1, time=frame)

  # Update the line object
  line.set_data(x, y)

  # Return the line object
  return (line,)
```

```python

animation = FuncAnimation(fig, update, interval=50, blit=True)

# Show the plot
#plt.show()
from IPython.display import HTML
HTML(animation.to_jshtml())
```
